```python
#payment vs tip
credit_tip=tax['tip'][tax['payment']=='credit card']
cash_tip=tax['tip'][tax['payment']=='cash']
from scipy import stats
t_stat,p_val=stats.ttest_ind(credit_tip,cash_tip)
print("independent T-Test statistic:",t_stat)
print("p-value:",p_val)
alpha=0.05

if(p_val<alpha):
    print("Reject null hypothesis: The tip for credit card payment is different from cash")
else:
    print("Fail to reject null hypothesis")
```

```python
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
print(tax.columns)
```

```python
cash = tax[tax['payment'] == 'cash']['fare']
card = tax[tax['payment'] == 'credit card']['fare']

f_stat, p_value = stats.f_oneway(cash, card)
print(f_stat, p_value)
```

```python
model = ols('fare ~ C(payment)', data=tax).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
tax_tukey = tax[['tip', 'pickup_borough']].copy()

tax_tukey['tip'] = pd.to_numeric(tax_tukey['tip'], errors='coerce')
tax_tukey['pickup_borough'] = tax_tukey['pickup_borough'].astype(str)

tax_tukey = tax_tukey.dropna()

tax_tukey = tax_tukey[tax_tukey['pickup_borough'].str.strip() != ""]
tax_tukey.reset_index(drop=True, inplace=True)

tukey = pairwise_tukeyhsd(
    endog=tax_tukey['tip'].values,
    groups=tax_tukey['pickup_borough'].values,
    alpha=0.05
)
print(tukey)
```

```python
# ANOVA Test
#regions = [group['Spending_Score'].values for name, group in df.groupby('pickup_borough')]
regions = [group['tip'].values for name, group in tax.groupby('pickup_zone')]
f_stat, p_val = stats.f_oneway(*regions)
print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)
 
if p_val < 0.05:
    print("Reject null hypothesis: tip differs by pickup zone")
else:
    print("Fail to reject null hypothesis: No significant difference by pickup zone")
```

```python
# ANOVA Test
#regions = [group['Spending_Score'].values for name, group in df.groupby('pickup_borough')]
regions = [group['tip'].values for name, group in tax.groupby('dropoff_zone')]
f_stat, p_val = stats.f_oneway(*regions)
print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)
 
if p_val < 0.05:
    print("Reject null hypothesis: tip differs by dropoff zone")
else:
    print("Fail to reject null hypothesis: No significant difference by dropoff zone")
```
