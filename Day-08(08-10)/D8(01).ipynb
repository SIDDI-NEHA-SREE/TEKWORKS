```python
import numpy as np
Arr=np.arange(10,50)
print(Arr)
print()
Arr1=np.random.rand(3,3)
print(Arr1)
print()
```

```python
#1
print("Array attributes:")
print("Size of array 1: ",np.size(Arr))
print("Size of array 2: ",np.size(Arr1))
print("Shape of array 1: ",np.shape(Arr))
print("Shape of array 2: ",np.shape(Arr1))
print("Data type of array 1: ",Arr.dtype)
print("Data type of array 2: ",Arr1.dtype)
```

```python
#2
print("Basic Operations")
Arr=Arr+5
print("Added 5 to every element: ",Arr)
Arr2=np.random.rand(3,3)
print("Multiplication of two arrays:\n ",Arr1*Arr2)
```

```python
#3
Arr3=np.arange(1,26)
Arr3=Arr3.reshape(5,5)
print(Arr3)
print("First row: ",Arr3[0])
print("Last column: ",Arr3[:,-1])
print("Subarray: \n",Arr3[:3,:2])
```

```python
#4
Arr4=np.random.randint(10,100,10)
print(Arr4)
print("Maximum value: ",np.max(Arr4))
print("Minimum value: ",np.min(Arr4))
print("Mean: ",np.mean(Arr4))
print("Standard deviation: ",np.std(Arr4))
```
